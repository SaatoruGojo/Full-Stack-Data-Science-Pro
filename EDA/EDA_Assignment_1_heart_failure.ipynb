# About - Dataset

The dataset provided for this project is focused on heart failure, a critical medical condition characterized by the heart's inability to pump
blood effectively, leading to inadequate circulation throughout the body. Here's an overview of the dataset:

# Title: Heart Failure Clinical Records Dataset

`Source:` The dataset was collected at the Faisalabad Institute of Cardiology and the Allied Hospital in Faisalabad, Punjab, Pakistan, during the period of April to December in 2015. It was later made available on Kaggle

`Description:` The dataset comprises medical records of 299 patients diagnosed with heart failure. These records include various clinical, body, and lifestyle information of the patients. The dataset encompasses 13 features, providing insights into the patients' health status and potential risk factors for heart failure

# Features:

`Age :` The age of the patient (numeric).

`Anaemia :` Indicates the presence of anemia (binary: 0 for no, 1 for yes).
`Creatinine Phosphokinase (CPK) :` Level of creatinine phosphokinase enzyme in the blood (numeric).

`Diabetes :` Indicates whether the patient has diabetes (binary: 0 for no, 1 for yes).

`Ejection Fraction :` Percentage of blood leaving the heart at each contraction (numeric).

`High Blood Pressure :` Indicates whether the patient has high blood pressure (binary: 0 for no, 1 for yes).

`Platelets :` Platelet count in the blood (numeric).

`Serum Creatinine :` Level of creatinine in the blood (numeric).

`Serum Sodium :` Level of sodium in the blood (numeric).

`Sex :` Gender of the patient (binary: 0 for female, 1 for male).

`Smoking :` Indicates whether the patient smokes (binary: 0 for no, 1 for yes).

`Time :` Follow-up period (in days) for the patient's condition (numeric).

`Death Event :` indicates whether the patient died during the follow-up period (binary: 0 for no, 1 for yes)

# 1- What is the distribution of age among heart failure patients in the dataset?

```python
# Distribution of age
plt.figure(figsize=(10, 6))
plt.hist(data['age'], bins=20, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.title('Distribution of Age among Heart Failure Patients')
plt.show()
```

```python
# Observation:
# - The distribution of age in the dataset is right-skewed, meaning that there are more younger patients than older patients.
# - The majority of patients are between 40 and 70 years old.
# - There are a few outliers, with some patients being over 90 years old.
```

#  2- How does the death rate vary with age?

```python
# Death rate by age
age_death_rate = data.groupby('age')['DEATH_EVENT'].mean()
plt.figure(figsize=(12, 6))
sns.lineplot(x=age_death_rate.index, y=age_death_rate.values)
plt.xlabel('Age')
plt.ylabel('Death Rate')
plt.title('Death Rate by Age')
plt.show()
```

```python
# Observation:
# - The death rate increases with age.
# - The death rate is highest for patients over 80 years old.
# - There is a slight decrease in the death rate for patients in their 70s,
```

# 3-What is the percentage of male and female patients in the dataset?

```python
# Gender distribution
gender_counts = data['sex'].value_counts(normalize=True) * 100
plt.figure(figsize=(10, 6))
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette='viridis')
plt.xticks([0, 1], ['Female', 'Male'])
plt.xlabel('Gender')
plt.ylabel('Percentage of Patients')
plt.title('Percentage of Male and Female Patients')
plt.show()
```

# 4 . How does platelets count varies among different age group?

```python
# Platelet count by age group
data['age_group'] = pd.cut(data['age'], bins=[0, 40, 50, 60, 70, 80, 90, 100])
plt.figure(figsize=(12, 6))
sns.boxplot(x='age_group', y='platelets', data=data)
plt.xlabel('Age Group')
plt.ylabel('Platelet Count')
plt.title('Platelet Count by Age Group')
plt.show()
```

```python
# Observation:
# - There is a general trend of decreasing platelet count with increasing age group.
# - There are a few outliers in each age group, but they are more common in the older age groups.
```

# 5. Is there a correlation between  creatinine and sodium levels in blood ?


```python
# Correlation between creatinine and sodium
plt.figure(figsize=(10, 6))
sns.scatterplot(x='serum_creatinine', y='serum_sodium', data=data)
plt.xlabel('Serum Creatinine')
plt.ylabel('Serum Sodium')
plt.title('Correlation between Serum Creatinine and Serum Sodium')
plt.show()

correlation = data[['serum_creatinine', 'serum_sodium']].corr()
print(correlation)
```

```python
# Observation:
# - There is a positive correlation between serum creatinine and death rate.
# - This means that patients with higher serum creatinine levels are more likely to die.
# - This is likely because serum creatinine is a marker of kidney function, and decreased kidney function is associated with increased mortality.
```

# 6.How does prevalance of high blood pressure differ between male and female patient?

```python
# High blood pressure by gender
high_bp_gender = data.groupby('sex')['high_blood_pressure'].mean() * 100
plt.figure(figsize=(10, 6))
sns.barplot(x=high_bp_gender.index, y=high_bp_gender.values, palette='viridis')
plt.xticks([0, 1], ['Female', 'Male'])
plt.xlabel('Gender')
plt.ylabel('Percentage with High Blood Pressure')
plt.title('Prevalence of High Blood Pressure by Gender')
plt.show()
```

# 7.what is the relationship between smoking habits and occurance of heart failures ?

```python
# Smoking habits and heart failure
smoking_heart_failure = data.groupby('smoking')['DEATH_EVENT'].mean() * 100
plt.figure(figsize=(10, 6))
sns.barplot(x=smoking_heart_failure.index, y=smoking_heart_failure.values, palette='viridis')
plt.xticks([0, 1], ['Non-Smokers', 'Smokers'])
plt.xlabel('Smoking Status')
plt.ylabel('Death Rate (%)')
plt.title('Death Rate by Smoking Status')
plt.show()
```

# 8.Are there any noticable pattern in distrubution of death event across different age group ?

```python
# Death events by age group
plt.figure(figsize=(12, 6))
sns.histplot(data=data, x='age', hue='DEATH_EVENT', multiple='stack', bins=20)
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.title('Death Events by Age Group')
plt.show()
```

# 9.Is there any signifiacant differance in ejection fraction between patient and without diabetes?


```python
# Ejection fraction by diabetes status
plt.figure(figsize=(12, 6))
sns.boxplot(x='diabetes', y='ejection_fraction', data=data)
plt.xlabel('Diabetes Status')
plt.ylabel('Ejection Fraction')
plt.title('Ejection Fraction by Diabetes Status')
plt.show()
```

# 10. How does the serum creatinine level vary between patients who survived and who didn't?

```python
# Serum creatinine by survival status
plt.figure(figsize=(12, 6))
sns.boxplot(x='DEATH_EVENT', y='serum_creatinine', data=data)
plt.xlabel('Death Event')
plt.ylabel('Serum Creatinine')
plt.title('Serum Creatinine by Survival Status')
plt.show()
```

```python
# Observation:
# - Patients who died have significantly higher serum creatinine levels compared to patients who survived'
# - This suggests that elevated serum creatinine levels are associated with an increased risk of death in patients with heart failure.
```
